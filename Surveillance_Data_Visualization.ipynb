```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/updated_combined_operations_2023-2024.csv')

# 1. Create a vertical bar chart from "Category" column for the different categories of incident (Taxonomy)
plt.figure(figsize=(8, 6))
df['Category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number Per Category')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# 2. Create a pie chart from "Type" as there are 02 types of incidents Events( no security aspects but may require security protection)
# and Incidents(security related ones)
plt.figure(figsize=(5, 5))
df['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['tomato', 'gold'])
plt.title('Type of Incidents')
plt.ylabel('')  # Hide y-axis label for clarity
plt.tight_layout()
plt.show()
```

```python
# 3. Create a horizontal bar chart from "Action taken by JTFGC"
# JTFGC stands for JOINT TASK FORCE OF GONDWANA CITY, the JTFGC is also the authorities of the operation center which
# several securiy patrols dispatched within the city and ready to intervene. The JTFGC takes action after viewing the surveillance video (live
# or post facto)
plt.figure(figsize=(8, 6))
df['Action taken by JTFGC'].value_counts().plot(kind='barh', color='lightgreen')
plt.title('Immediate Action Taken by JTFGC')
plt.xlabel('Number of Actions')
plt.tight_layout()
plt.show()
```

```python
# 4. Count occurrences of each sensor, these are the cameras installed various
#locations of the Gondwana city and operated by operators 24/7 inside the Tactical Operation
# Center (TOC)
sensor_counts = df['Sensors used'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
sensor_counts.plot(kind='pie', autopct='%1.1f%%', colormap='Pastel1')
plt.title('Cameras Usage on Events/Incidents')
plt.ylabel('')  # Hide the y-label
plt.tight_layout()
plt.show()
```

```python
#5.  Count occurrences per detector, These are the names of the operator for the purpose of individual KPI.
#However, JTFGC in this case stand for the information received from an Officer of the Headquarter of the JTFGC
# Planned is for the event scheduled by Fragmented order or operation order.
detected_counts = df['Detected by'].value_counts()

# Create a horizontal bar chart
plt.figure(figsize=(8, 4))
detected_counts.plot(kind='barh', color='purple')
plt.title('Events/Incidents Detected by')
plt.xlabel('Number of Events/Incidents')
plt.ylabel('Detector')
plt.tight_layout()
plt.show()
```
